# Weight and Height

```python
df.Weight.unique()
```

```python
wList = [] # Keeping 'kg' as our unit

for val in df.Weight.values:
    val = str(val)
    if 'kg' in val:
        #print(val)
        #print(val[:-2])
        wList.append(float(val[:-2]))
    elif 'lbs'in val:
        wList.append(float(val[:-3])*0.453592) # converting lbs to kg
    else:
        wList.append(val)
```

```python
wList
```

```python
dfNew['wt_kg'] = wList
```

```python
dfNew
```

```python
df.Height.unique()
```

```python
tempList = []

for val in df.Height:
    if 'cm' not in val:
        tempList.append(val)
        
xx = list(set(tempList))
print(xx)
```

```python
hgtList = []

for val in df.Height.values:
    val = str(val)
    if 'cm' in val:
        ht = float(val[:-2])
        hgtList.append(ht)
    elif val in xx:
        ft = val[0]
        inch = val[-2]
        # converting to cm
        hgt1 = float(ft)*30.48 + float(inch)*2.54
        hgtList.append(hgt1)
    else:
        hgtList.append(val)
```

```python
hgtList
```

```python
TASK 5
# Value, Wage and Release Clause columns: convert to Float 
```

```python
df.Value.unique()
```

```python
valList = [] # thousands

for val in df.Value:
    if 'K' in val:
        val = float(val[1:-1])*1000
        valList.append(val)
    elif 'M' in val:
        val = float(val[1:-1])*1000000
    else:
        valList.append(float(0))
```

```python
valList
```

```python
df.Wage.unique()
```

```python
wageList = []

for w in df.Wage:
    if 'K' in w:
        w = w[1:-1]
        wageList.append(float(w)*1000)
    else:
        w = w[1:]
        wageList.append(float(w)*1000)
```
